# Combining Data Frames

#### Import Pandas module

```python
import pandas as pd
```

#### Load csv files into Pandas Data Frame

```python
data_2015 = pd.read_csv("World Happiness Data 2015.csv")
data_2016 = pd.read_csv("World Happiness Data 2016.csv")
data_2017 = pd.read_csv("World Happiness Data 2017.csv")
```

```python
data_2016
```

```python
data_2017
```

#### Adding column with respective year value to each DataFrame 

```python
data_2015["Year"] = 2015
data_2016["Year"] = 2016
data_2017["Year"] = 2017
```

```python
data_2015[["Year"]]
```

### Extracting same number of columns in two DataFrames with five values each

```python
column_2015 = data_2015[["Country","Happiness Score","Year"]].head()
column_2016 = data_2016[["Country","Happiness Score","Year"]].head()
```

```python
column_2015
```

```python
column_2016
```

#### Joining the DataFrames in rows

```python
pd.concat([column_2015,column_2016], axis = 0, ignore_index = True)
```

#### Joining the DataFrames in columns

```python
pd.concat([column_2015,column_2016], axis = 1)
```

### Renaming column of "data_2017" for joining DataFrame

```python
data_2017.rename({"Happiness.Score":"Happiness Score"},axis = 1, inplace = True)
```

```python
data_2017
```

### Extracting same number of columns in all DataFrames with five values each

```python
column_2015 = data_2015[["Country","Happiness Score","Year"]].head()
column_2016 = data_2016[["Country","Happiness Score","Year"]].head()
column_2017 = data_2017[["Country","Happiness Score","Year"]].head()
```

#### Joining the DataFrames in rows

```python
pd.concat([column_2015, column_2016, column_2017], axis = 0, ignore_index = True)
```

#### Joining the DataFrames in columns

```python
pd.concat([column_2015, column_2016, column_2017], axis = 1)
```

# Merging DataFrames

```python
column_2015
```

```python
column_2016
```

### inner 

```python
pd.merge(left = column_2015, right = column_2016, how = "inner", on = "Country", suffixes=(" 2015", " 2016"))
```

### left

```python
pd.merge(left = column_2015, right = column_2016, how = "left", on = "Country", suffixes=(" 2015", " 2016"))
```

### right

```python
pd.merge(left = column_2015, right = column_2016, how = "right", on = "Country", suffixes=(" 2015", " 2016"))
```
